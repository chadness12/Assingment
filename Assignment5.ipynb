```python
class Counter():
    def __init__(self, count, step=1): # 생성자
        self.count = count
        self.step = step
    def __str__(self): #  특수 함수 객체의 문자열 표현을 돌려주기로 되어있습니다.
        return '[Count (step: {})] {}'.format(self.step, self.count)  
    def incr(self):
        self.count = self.count+self.step
    def __call__(self): # __call__은 클래스의 객체가 함수처럼 호출되면 실행되는 함수이다.
        self.count = self.count+self.step
    #def __add__(self, another):        
        #return self.count + another    -> 이렇게 하면 self.count+anoter인 int만 반환됨
    def __add__(self, another): # __add__
        self.count=self.count + another
        return self
    def __sub__(self , another):
        self.count=self.count - another
        return self
    def __lt__(self, another): # < 비교를 위해
        return self.count < another 
    
    def __le__(self, another): # <=비교를위해
        return self.count <= another
    
    def __eq__(self, another): # ==비교를위해
        return self.count == another
    
    def __ne__(self, another): # != 비교를 위해
        return self.count != another 
    
    def __gt__(self, another): # > 비교를 위해
        return self.count > another 
     
    def __ge__(self, another): # >= 비교를 위해
        return self.count >= another 
    
     
c = Counter(10)
d = Counter(10, 2)
print(c)
print(d)
c.incr()
d.incr()
print(c)
print(d)
c()
d()
print(c)
print(d)
c = c + 5
d = d - 5
print(c)
print(d)
print(c > 10)
print(d > 10)
print(c < 10)
print(d < 10)
print(c == 17)
print(d != 9)   
    
    
        
        
```

### 4번 문제
다음은 내장 자료형 list를 서브클래싱하여 만든 MySet 클래스 정의 내용이다. 다음 클래스 정의에서 __init__(), __str()__(), elimicate_duplicate()의 세 개의 메소드 코드 내용을 자신이 다른 사람에게 가르친다고 생각하며 설명해보시오.

* MySet은 집합(Set) 자료형을 정의하려는 의도하에 만들어진 클래스이다.

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)

    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print(s)
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print(t)
```

__ init __ 은 이미 만들어져 있는 메소드로 변수 초기화 할 용도로 사용한다. s = MySet([1, 2, 2, 3])을 선언 할시 init에서는 이것을 받아 자기자신에 넣어주고 중복을 제거한다.

__ str __ 은 출력을 위해 사용한다  print(s)를 위한 메소드 이다. result에 자기자신 안에 있는 인자를 문자열 result에 더한후 return 한다.

마지막으로 @staticmethod  정적 메소드를 뜻한다. 하지만 파이썬에서는 정적 메소드도 인서턴스에서도 접근이 가능하다. 이 함수의 역할은 중복을 제거하는 함수이다. e가 s에 없다면 중복이아니므로 s에 값을 넣어주고 최종적으로 list l에 중복 제거한 s값을 넣어준다

### 5번 문제

* 정답을 각 요구사항별로 입력할 필요 없이 요구사항 3개 전체에 대해 1개의 MySet 클래스 정의 코드를 제시하고 주석등으로 해당 코드를 설명한다.
* 요구사항 1. | 연산으로 두 집합의 합집합을 반환한다.

> u = s | t   
> print(u)

* 요구사항 2. & 연산으로 두 집합의 교집합을 반환한다.

> u = s & t   
> print(u)

* 요구사항 3. - 연산으로 두 집합의 차집합을 반환한다.

> s = MySet([1, 2, 3])   
> t = MySet([3, 4, 5])   
> u = s - t   
> print(u)   
> MySet: {1, 2}   

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result
    def __or__ (self,another): # | 합집합 연산
        save = self + another
        MySet.eliminate_duplicate(save)
        # save = self.I + another 로 해야할 줄 알았는데, __init__에서 self.l이라는 것을 만들어주지않음
        for e in self: self.remove(e)
        for e in save: self.append(e)
        MySet.eliminate_duplicate(self)# self내 중복 제거
        self.sort()
        return self
    
    def __and__(self,another): # & 교집합 연산
        save = []
        for e in self:
            for i in another:
                if e==i: save.append(e) # 중복된다면 append
        for e in self: self.remove(e)
        for e in save: self.append(e)
        MySet.eliminate_duplicate(self)# self내 중복 제거
        return self
    
     
    def __sub__(self,another): # - 차집합 연산
        save = []
        for e in self:
            for i in another:
                if e==i: save.append(e)
        for e in self:
            for i in save:
                if e==i: self.remove(i) # self 내 another과의 중복값을 지워줌
        return self
    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)

    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    u = s | t
    print(u)
    s = MySet([1, 2, 2, 3])
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    u = s & t
    print(u)
    s = MySet([1, 2, 3])
    t = MySet([3, 4, 5])
    u = s - t
    print(u)
```
