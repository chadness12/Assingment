        s = '  first star   :   second star   :    third star  '
        l = ['first star', 'second star', 'third star']
* 1) for ~ in 문을 이용한 코드를 작성하시오.
* 2) 리스트 내포(list comprehension)를 이용한 코드를 작성하시오.

```python
# for~ in 문 활용 코드
s = '  first star   :   second star   :    third star  '
l=s.split(':')
for i in range (0,len(l)):
    l[i]=l[i].strip()
print(l)
```

:를 기준으로 잘라 l에 넣고 l의 길이 만큼 l[i]에 양쪽 공백을 지워넣고 출력합니다

### 문제 5번
다음과 같이 0보다 큰 정수 리스트 변수 list를 인자로 받는 함수 addall(list)와 addallodd(list)를 for ~ in 리터널과 리스트 내포 방식으로 각각 제시하시오.
* 1) addall(list) 함수 (리스트 내의 모든 정수를 더하는 함수로서 해답으로는 for ~ in 리터럴과 리스트 내포 방식으로 각각 제시하시오.)
        + >>> addall([1])
          1
        + >>> addall([1, 2, 3, 4, 5, 6, 7, 8, 9])
         45
* 2) addallodd(list) 함수 (리스트내의 모든 홀수를 더하는 함수로서 해답으로는 for ~ in 리터럴과 리스트 내포 방식으로 각각 제시하시오.)
        + >>> addallodd([1])
         1
        + >>> addallodd([1, 2, 3, 4, 5, 6, 7, 8, 9])
         25
* [참고 1]: 리스트 내포 방식으로 addall() 및 addallodd()함수를 만들 때엔 리스트 내포에 의해 생성된 리스트 결과에 대해 내장 함수를 사용하는 방식을 고려해보시오 (주의: 리스트 내포 방식의 addall() 및 addallodd() 함수의 몸체는 단 1줄로 작성되어야 한다.)
* [참고 2]: sum() 등의 내장함수 사용 가능함

```python
# 리스트 내포 방식 활용
def addall(list):
    print(sum(list))
a = [x for x in range(0,10)]
addall(a)
```

a에 리스트 내포함수를 통해 값을 넣고 sum을 통해 모든값을 합칩니다.

```python
# for~ in 방식 활용
def addall(list):
    sum = 0
    for i in range(len(list)):
        sum = sum + list[i]
    print(sum)
s = []
for i in range(10):
    s.append(i)
addall(s)
```

s.append(i)를 통해 0~9까지의 값을 넣습니다. addall함수는 sum+list[i]를 통해 하나씩 값을 더해가고 마지막에 출력하는함수입니다.

```python
# 리스트 내포 방식 활용
def addoddall(list):
    print(sum(x for x in list if x%2))
a = [x for x in range(0,10)]
addoddall(a)
```

리스트 내포에 조건을 활용하여 x가 2로 나눠지지 않을시 실행되도럭 했습니다. 홀수를 모두 더해 출력합니다

```python
# for~ in 방식 활용
def addoddall(list):
    sum = 0
    for i in range(len(list)):
        if(list[i]%2==1):
            sum=sum+list[i]
    print(sum)
a = []
for i in range(10):
    a.append(i)
addoddall(a)
```

for~in 에 if문을 추가하여 2로나눈 나머지가 1일시 sum에 list값을 추가하는 코드입니다. 홀수만 더해져 출력됩니다.
